```python
import pandas as pd 
import numpy as np

df_ten = pd.read_excel("result.xlsx",sheet_name="10 кВ")
df_little = pd.read_excel("result.xlsx",sheet_name="0.38 кВ")
df_sum = pd.read_excel("result.xlsx",sheet_name="Cуммарные")
```

```python
def month_counter(month,year):
    if month in [1,3,5,7,8,10,12]:
        return 31
    elif month == 2 and (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return 29
    else:
        return 30
    
```

Фактически расчет лапками числа оборудования по справочным данным для 10 кВ

```python
df_ten["number_DGR"] = [round(df_ten["dWDGR"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*2.89)/8760)) for i in range(len(df_ten["dWDGR"]))]
df_ten["number_TOR"] = [round(df_ten["dWTOR"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.1)/8760)) for i in range(len(df_ten["dWTOR"]))]
df_ten["number_KU"] = [round(df_ten["dWKU"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.1)/8760)) for i in range(len(df_ten["dWKU"]))]
df_ten["number_CK"] = [round(df_ten["dWCK"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.1)/8760)) for i in range(len(df_ten["dWCK"]))]
df_ten["number_VR"] = [round(df_ten["dWVR"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.021)/8760)) for i in range(len(df_ten["dWVR"]))]
df_ten["number_OPN"] = [round(df_ten["dWOPN"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.001)/8760)) for i in range(len(df_ten["dWOPN"]))]
df_ten["number_TT"] = [round(df_ten["dWTrT"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.1)/8760)) for i in range(len(df_ten["dWTrT"]))]
df_ten["number_TN"] = [round(df_ten["dWTrN"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*1.9)/8760)) for i in range(len(df_ten["dWTrN"]))]
df_ten["number_shet"] = [round(df_ten["dWshet"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.1)/8760)) for i in range(len(df_ten["dWshet"]))]
df_ten["number_spps"] = [round(df_ten["dWspps"][i]/((month_counter(df_ten["month"][i],df_ten["year"][i])*24*0.1)/8760)) for i in range(len(df_ten["dWspps"]))]
```
