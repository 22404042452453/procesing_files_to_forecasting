```python
import os 
import pandas as pd 
```

```python
def checker(text):
    if ',' in text:
        text = text.replace(",",".")
    else:
        pass
    
    return text.strip()
```

```python
def read_files(path):
    sum_list, ten_list, six_list, little_list = [],[],[],[]
    
    f = open(path,'r')
    
    for number,text in enumerate(f.readlines()):
        text = checker(text)

        if number == 4 or number == 5:
            sum_list.append(int(text))
            ten_list.append(int(text))
            six_list.append(int(text))
            little_list.append(int(text))

        if 6<=number<=34:
            if number == 9:
                sum_list.append(float(text))
                sum_list+=[0,0]
            elif number == 23:
                sum_list.append(float(text))
                sum_list+=[0]

            else:
                sum_list.append(float(text))

        elif 35<=number<=63:
            if number == 38:
                ten_list.append(float(text))
                ten_list+=[0,0]
            elif number == 52:
                ten_list.append(float(text))
                ten_list+=[0]
            else:
                ten_list.append(float(text))

        elif 64 <= number <= 92:
            if number == 67:
                six_list.append(float(text))
                six_list+=[0,0]
            elif number == 81:
                six_list.append(float(text))
                six_list+=[0]

            else:
                six_list.append(float(text))
        else:
            if number > 92:
                if number == 98:
                    little_list.append(float(text))
                    little_list+= [0,0,0,0,0]
                elif number == 99:
                    little_list.append(float(text))
                    little_list+= [0,0,0]
                elif number == 104:
                    little_list.append(float(text))
                    little_list+= [0]
                elif number == 105:
                    little_list.append(float(text))
                    little_list+= [0]
                else:            
                    little_list.append(float(text))
            else:
                continue
    f.close()
    
    return sum_list, ten_list, six_list, little_list
       
        
```
